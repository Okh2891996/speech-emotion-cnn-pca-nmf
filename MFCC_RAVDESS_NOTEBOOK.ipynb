# REVDESS TRAIN TEST TESS

```python
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import regularizers
from sklearn.model_selection import train_test_split
import random

# ── PARAMETERS ──────────────────────────────────────────────────────────────
ravdess_dir = r"C:\Users\okh96\OneDrive - Toronto Metropolitan University (TMU)\TMU_D2L\EE8104 - Adaptive Signal Processing\Assignment5\ravdess_by_emotion"

sr = 16000        # Sampling rate
n_mfcc = 13       # Number of MFCCs
max_length = 85   # Number of frames in MFCC

# ── EMOTION LABEL MAPPING ───────────────────────────────────────────────────
emotion_map = {
    'angry': 0,
    'disgust': 1,
    'fear': 2,
    'happy': 3,
    'neutral': 4,
    'pleasant_surprised': 5,  # handle alternate spelling
    'sad': 6
}

# ── MFCC EXTRACTION FUNCTION ────────────────────────────────────────────────
def extract_mfcc(file_path, max_length=85):
    y, _ = librosa.load(file_path, sr=sr)
    hop_length = int(len(y) / (max_length - 1))
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)
    if mfcc.shape[1] < max_length:
        pad_width = max_length - mfcc.shape[1]
        mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='constant')
    elif mfcc.shape[1] > max_length:
        mfcc = mfcc[:, :max_length]
    return mfcc

# ── GENERAL DATA LOADER FUNCTION FOR RAVDESS ────────────────────────────────
def load_all_data(base_dir):
    data = []
    labels = []
    for emotion in os.listdir(base_dir):
        emotion_path = os.path.join(base_dir, emotion)
        if not os.path.isdir(emotion_path):
            continue
        label = emotion_map.get(emotion.lower())
        if label is None:
            print(f"Skipping unknown emotion folder: {emotion}")
            continue
        for file in os.listdir(emotion_path):
            file_path = os.path.join(emotion_path, file)
            try:
                mfcc = extract_mfcc(file_path, max_length)
                mfcc = np.expand_dims(mfcc, axis=-1)
                data.append(mfcc)
                labels.append(label)
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
    return np.array(data), np.array(labels)

# ── LOAD AND SPLIT RAVDESS DATA ─────────────────────────────────────────────
X_all, y_all = load_all_data(ravdess_dir)

# Shuffle and split into train, val, test (70/15/15)
X_temp, X_test, y_temp, y_test = train_test_split(X_all, y_all, test_size=0.15, random_state=42, stratify=y_all)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.1765, random_state=42, stratify=y_temp)  # 0.1765 * 0.85 ≈ 0.15

print("Train data shape:", X_train.shape)
print("Validation data shape:", X_val.shape)
print("Test data shape:", X_test.shape)

# ── DEFINE CNN MODEL ────────────────────────────────────────────────────────
class EmotionClassifier(tf.keras.Model):
    def __init__(self, num_classes):
        super(EmotionClassifier, self).__init__()
        self.weight_decay = 1e-4
        
        # First Conv Block
        self.conv1 = layers.Conv2D(32, 3, padding='same',
                                 kernel_regularizer=regularizers.l2(self.weight_decay))
        self.bn1 = layers.BatchNormalization()
        self.act1 = layers.Activation('relu')
        self.pool1 = layers.MaxPooling2D(2)
        self.drop1 = layers.Dropout(0.3)
        
        # Second Conv Block
        self.conv2 = layers.Conv2D(64, 3, padding='same',
                                 kernel_regularizer=regularizers.l2(self.weight_decay))
        self.bn2 = layers.BatchNormalization()
        self.act2 = layers.Activation('relu')
        self.pool2 = layers.MaxPooling2D(2)
        self.drop2 = layers.Dropout(0.3)
        
        # Third Conv Block
        self.conv3 = layers.Conv2D(128, 3, padding='same',
                                 kernel_regularizer=regularizers.l2(self.weight_decay))
        self.bn3 = layers.BatchNormalization()
        self.act3 = layers.Activation('relu')
        self.pool3 = layers.MaxPooling2D(2)
        self.drop3 = layers.Dropout(0.3)
        
        # Dense Layers
        self.flatten = layers.Flatten()
        self.dense1 = layers.Dense(128, activation='relu',
                                kernel_regularizer=regularizers.l2(self.weight_decay))
        self.drop4 = layers.Dropout(0.4)
        self.dense2 = layers.Dense(num_classes, activation='softmax')

    def call(self, inputs, training=False):
        # First Conv Block
        x = self.conv1(inputs)
        x = self.bn1(x, training=training)
        x = self.act1(x)
        x = self.pool1(x)
        x = self.drop1(x, training=training)
        
        # Second Conv Block
        x = self.conv2(x)
        x = self.bn2(x, training=training)
        x = self.act2(x)
        x = self.pool2(x)
        x = self.drop2(x, training=training)
        
        # Third Conv Block
        x = self.conv3(x)
        x = self.bn3(x, training=training)
        x = self.act3(x)
        x = self.pool3(x)
        x = self.drop3(x, training=training)
        
        # Dense Layers
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.drop4(x, training=training)
        return self.dense2(x)

# ── COMPILE AND TRAIN ───────────────────────────────────────────────────────
num_classes = len(emotion_map)

model = EmotionClassifier(num_classes)
model.build(input_shape=(None, n_mfcc, max_length, 1))  # define input shape
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Optional: TensorBoard
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir='logs/emotion_classifier', histogram_freq=1)

model.fit(X_train, y_train,
          epochs=350,
          validation_data=(X_val, y_val),
          callbacks=[tensorboard_callback])

# ── EVALUATE ────────────────────────────────────────────────────────────────
test_loss, test_acc = model.evaluate(X_test, y_test)
print("Test loss:", test_loss)
print("Test accuracy:", test_acc)
```

```python
# ── TEST ON FULL RAVDESS-BY-EMOTION FOLDER ───────────────────────────────────
# Re-load full dataset (ensure same pre-processing as before)
tess_dir = "./TESSdataset/"
X_full, y_full = load_all_data(tess_dir)

# Predict
y_pred_probs = model.predict(X_full)
y_pred = np.argmax(y_pred_probs, axis=1)

# Calculate accuracy
accuracy = np.mean(y_pred == y_full)
print("Full tess dataset accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Classification report
print(classification_report(y_full, y_pred, target_names=emotion_map.keys()))

# Confusion matrix
cm = confusion_matrix(y_full, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=emotion_map.keys(), yticklabels=emotion_map.keys())
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix on Full tess Dataset")
plt.show()
```

# NEW Training process

```python
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras import layers, regularizers
from sklearn.model_selection import train_test_split
import random

# ── PARAMETERS ──────────────────────────────────────────────────────────────
ravdess_dir = r"C:\Users\okh96\OneDrive - Toronto Metropolitan University (TMU)\TMU_D2L\EE8104 - Adaptive Signal Processing\Assignment5\ravdess_by_emotion"

sr = 16000        # Sampling rate
n_mfcc = 13       # Number of MFCCs
max_length = 85   # Number of frames in MFCC

# ── EMOTION LABEL MAPPING ───────────────────────────────────────────────────
emotion_map = {
    'angry': 0,
    'disgust': 1,
    'fear': 2,
    'happy': 3,
    'neutral': 4,
    'pleasant_surprised': 5,
    'sad': 6
}

# ── MFCC AUGMENTATION ───────────────────────────────────────────────────────
def augment_mfcc(mfcc):
    # time‑mask
    t = mfcc.shape[1]
    t0 = np.random.randint(0, t//4)
    w  = np.random.randint(0, t//8)
    mfcc[:, t0:t0+w] = 0
    # freq‑mask
    f = mfcc.shape[0]
    f0 = np.random.randint(0, f//4)
    h  = np.random.randint(0, f//8)
    mfcc[f0:f0+h, :] = 0
    return mfcc

# ── MFCC EXTRACTION FUNCTION ────────────────────────────────────────────────
def extract_mfcc(file_path, max_length=85):
    y, _ = librosa.load(file_path, sr=sr)
    hop_length = int(len(y) / (max_length - 1))
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)
    if mfcc.shape[1] < max_length:
        pad = max_length - mfcc.shape[1]
        mfcc = np.pad(mfcc, ((0,0),(0,pad)), mode='constant')
    else:
        mfcc = mfcc[:, :max_length]
    return mfcc

# ── DATA LOADER ─────────────────────────────────────────────────────────────
def load_all_data(base_dir):
    X, y = [], []
    for emo in os.listdir(base_dir):
        p = os.path.join(base_dir, emo)
        if not os.path.isdir(p): continue
        lbl = emotion_map.get(emo.lower())
        if lbl is None: continue
        for f in os.listdir(p):
            if not f.lower().endswith('.wav'): continue
            fp = os.path.join(p, f)
            try:
                m = extract_mfcc(fp, max_length)
                X.append(np.expand_dims(m, -1))
                y.append(lbl)
            except:
                pass
    return np.array(X), np.array(y)

# ── LOAD & SPLIT ────────────────────────────────────────────────────────────
X_all, y_all = load_all_data(ravdess_dir)
X_temp, X_test, y_temp, y_test = train_test_split(
    X_all, y_all, test_size=0.15, random_state=42, stratify=y_all)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, random_state=42, stratify=y_temp)

print("Train:", X_train.shape, " Val:", X_val.shape, " Test:", X_test.shape)

# ── APPLY AUGMENTATION TO TRAINING SET ───────────────────────────────────────
X_train_aug = []
for m in X_train:
    mm = m[:,:,0]                      # remove channel dim
    mm = augment_mfcc(mm)             # augment
    X_train_aug.append(np.expand_dims(mm, -1))
X_train = np.array(X_train_aug)

# ── MODEL DEFINITION ────────────────────────────────────────────────────────
class EmotionClassifier(tf.keras.Model):
    def __init__(self, num_classes):
        super().__init__()
        wd = 1e-3
        # Conv Blocks with heavier dropout
        self.conv1 = layers.Conv2D(32,3,padding='same', kernel_regularizer=regularizers.l2(wd))
        self.bn1   = layers.BatchNormalization(); self.pool1 = layers.MaxPool2D()
        self.drop1 = layers.Dropout(0.5)
        
        self.conv2 = layers.Conv2D( 64,3,padding='same', kernel_regularizer=regularizers.l2(wd))
        self.bn2   = layers.BatchNormalization(); self.pool2 = layers.MaxPool2D()
        self.drop2 = layers.Dropout(0.5)
        
        self.conv3 = layers.Conv2D(128,3,padding='same', kernel_regularizer=regularizers.l2(wd))
        self.bn3   = layers.BatchNormalization(); self.pool3 = layers.MaxPool2D()
        self.drop3 = layers.Dropout(0.5)
        
        # Dense head
        self.flat   = layers.Flatten()
        self.dense1 = layers.Dense(128,activation='relu', kernel_regularizer=regularizers.l2(wd))
        self.drop4  = layers.Dropout(0.5)
        self.out    = layers.Dense(num_classes, activation='softmax')

    def call(self, x, training=False):
        x = self.pool1(self.bn1(self.conv1(x), training=training))
        x = self.drop1(x, training=training)
        x = self.pool2(self.bn2(self.conv2(x), training=training))
        x = self.drop2(x, training=training)
        x = self.pool3(self.bn3(self.conv3(x), training=training))
        x = self.drop3(x, training=training)
        x = self.flat(x)
        x = self.drop4(self.dense1(x), training=training)
        return self.out(x)

# ── COMPILE & TRAIN ────────────────────────────────────────────────────────
num_classes = len(emotion_map)
model = EmotionClassifier(num_classes)
model.build((None, n_mfcc, max_length, 1))
model.compile('adam', 'sparse_categorical_crossentropy', ['accuracy'])

tensorboard_cb = tf.keras.callbacks.TensorBoard(log_dir='logs/emotion_classifier', histogram_freq=1)
#earlystop_cb    = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(
    X_train, y_train,
    epochs=350,
    validation_data=(X_val, y_val),
    callbacks=[tensorboard_cb]
)

# ── EVALUATE ────────────────────────────────────────────────────────────────
loss, acc = model.evaluate(X_test, y_test)
print(f"Test loss: {loss:.4f}   Test accuracy: {acc:.4f}")
```

```python
# ── TEST ON FULL RAVDESS-BY-EMOTION FOLDER ───────────────────────────────────
# Re-load full dataset (ensure same pre-processing as before)
tess_dir = "./TESSdataset/"
X_full, y_full = load_all_data(tess_dir)

# Predict
y_pred_probs = model.predict(X_full)
y_pred = np.argmax(y_pred_probs, axis=1)

# Calculate accuracy
accuracy = np.mean(y_pred == y_full)
print("Full tess dataset accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Classification report
print(classification_report(y_full, y_pred, target_names=emotion_map.keys()))

# Confusion matrix
cm = confusion_matrix(y_full, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=emotion_map.keys(), yticklabels=emotion_map.keys())
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix on Full tess Dataset")
plt.show()
```
