```python
import os
import librosa
import numpy as np
import pandas as pd
from sklearn.decomposition import NMF
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

# ========= 1. Feature Extraction Function =========

def extract_mfcc(file_path, n_mfcc=13):
    try:
        y, sr = librosa.load(file_path, sr=None)
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        return np.mean(mfcc, axis=1)
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

ravdess_root = r".\ravdess_by_emotion"

ravdess_data = []

for emotion in os.listdir(ravdess_root):
        
    emotion_dir = os.path.join(ravdess_root, emotion)
    for filename in os.listdir(emotion_dir):
        if filename.endswith(".wav"):
            file_path = os.path.join(emotion_dir, filename)
            mfcc = extract_mfcc(file_path)
            if mfcc is not None:
                ravdess_data.append((mfcc, emotion))


combined_data = ravdess_data
X = np.array([x[0] for x in combined_data])
y = [x[1] for x in combined_data]

df = pd.DataFrame(X)
df['Emotion'] = y





# ========= 5. Normalize and Apply NMF =========

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df.drop(columns=['Emotion']))

nmf = NMF(n_components=5, init='random', random_state=42)
W = nmf.fit_transform(X_scaled)

W_df = pd.DataFrame(W, columns=[f'Component_{i+1}' for i in range(W.shape[1])])
W_df['Emotion'] = y

# ========= 6. Visualize Component Activation per Emotion =========

plt.figure(figsize=(10, 6))
sns.heatmap(W_df.groupby('Emotion').mean(), annot=True, cmap='viridis')
plt.title("NMF Component Activation per Emotion")
plt.ylabel("Emotion")
plt.xlabel("NMF Component")
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA on all utterances (no grouping)
pca = PCA(n_components=2)
pca_result_all = pca.fit_transform(W_df.drop('Emotion', axis=1))

# Add the PCA result to the original DataFrame
W_df['PC1'] = pca_result_all[:, 0]
W_df['PC2'] = pca_result_all[:, 1]

# Plot all data points colored by emotion
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.scatterplot(data=W_df, x='PC1', y='PC2', hue='Emotion', palette='tab10', s=60)
plt.title("PCA of All Utterances in NMF Space")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get the average NMF component activations per emotion
emotion_matrix = W_df.groupby('Emotion').mean()

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(emotion_matrix)

# Create a DataFrame for plotting
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'], index=emotion_matrix.index)

# Plot the emotions in 2D PCA space
plt.figure(figsize=(10, 7))
for emotion, (x, y) in pca_df.iterrows():
    plt.scatter(x, y, s=100)
    plt.text(x + 0.01, y + 0.01, emotion, fontsize=10)

plt.title("Emotion Clustering in NMF Component Space (via PCA)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
