```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import SVC
from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.utils import shuffle

#  PARAMETERS FOR FULLMAP MFCC 
SAMPLE_RATE   = 8000
DURATION      = 0.5
TARGET_LENGTH = int(SAMPLE_RATE * DURATION)
N_MFCC        = 13
MAX_FRAMES    = 85

# Paths for the two datasets
tess_dir    = "./TESSdataset"
ravdess_dir = "./ravdess_by_emotion"

#  FEATURE EXTRACTOR FOR 1385 MFCC MAP 
def extract_mfcc_flat(signal):
    hop_length = int(len(signal) / (MAX_FRAMES - 1))
    mfcc = librosa.feature.mfcc(
        y=signal,
        sr=SAMPLE_RATE,
        n_mfcc=N_MFCC,
        hop_length=hop_length
    )
    # pad or truncate to exactly (13, MAX_FRAMES)
    if mfcc.shape[1] < MAX_FRAMES:
        mfcc = np.pad(mfcc,
                      pad_width=((0, 0), (0, MAX_FRAMES - mfcc.shape[1])),
                      mode='constant')
    else:
        mfcc = mfcc[:, :MAX_FRAMES]

    return mfcc.flatten()  # (13*85)=1105-dim

#  DIRECTORY PROCESSING USING THE NEW EXTRACTOR 
def process_directory(directory):
    X, y = [], []
    for emotion in sorted(os.listdir(directory)):
        emotion_dir = os.path.join(directory, emotion)
        if not os.path.isdir(emotion_dir):
            continue
        for fname in sorted(os.listdir(emotion_dir)):
            path = os.path.join(emotion_dir, fname)
            try:
                sig, _ = librosa.load(path, sr=SAMPLE_RATE)
                sig = librosa.util.fix_length(sig, size=TARGET_LENGTH)
                sig = sig / (np.max(np.abs(sig)) + 1e-8)

                feats = extract_mfcc_flat(sig)
                X.append(feats)
                y.append(emotion.lower())
            except Exception as e:
                print(f"Skipped {path}: {e}")
    return np.vstack(X), np.array(y)

#  LOAD & COMBINE BOTH DATASETS 
print("Loading TESS dataset")
X_tess, y_tess = process_directory(tess_dir)
print("Loading RAVDESS dataset")
X_rav,  y_rav  = process_directory(ravdess_dir)

X, y = np.vstack((X_tess, X_rav)), np.concatenate((y_tess, y_rav))
X, y = shuffle(X, y, random_state=42)

#  TRAIN/TEST SPLIT 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

#  STANDARDIZE FOR PCA 
std_scaler  = StandardScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_test_std  = std_scaler.transform(X_test)

#  SHIFT TO NONNEGATIVE FOR NMF 
global_min = min(X_train.min(), X_test.min())
if global_min < 0:
    X_train_nn = X_train - global_min
    X_test_nn  = X_test  - global_min
else:
    X_train_nn = X_train.copy()
    X_test_nn  = X_test.copy()

#  MINMAX SCALE TO [0,1] 
mm_scaler  = MinMaxScaler(feature_range=(0,1))
X_train_mm = mm_scaler.fit_transform(X_train_nn)
X_test_mm  = mm_scaler.transform(X_test_nn)

#  CLIP ANY SMALL NEGATIVES BACK TO ZERO 
# (sometimes test features go slightly below the training min)
X_train_mm = np.clip(X_train_mm, 0, None)
X_test_mm  = np.clip(X_test_mm,  0, None)

#  NMF  5 components 
nmf         = NMF(n_components=2, random_state=42)
X_train_nmf = nmf.fit_transform(X_train_mm)
X_test_nmf  = nmf.transform(X_test_mm)

#  PCA  5 components 
pca         = PCA(n_components=2, random_state=42)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca  = pca.transform(X_test_std)

#  CONCATENATE REDUCED FEATURES 
X_train_reduced = np.hstack([X_train_nmf, X_train_pca])
X_test_reduced  = np.hstack([X_test_nmf,  X_test_pca])

#  TRAIN & EVALUATE SVM 
clf = SVC(kernel='rbf', C=10, gamma='scale', random_state=42)
clf.fit(X_train_reduced, y_train)

# Training performance
y_pred_train = clf.predict(X_train_reduced)
print(f"\nTraining Accuracy: {accuracy_score(y_train, y_pred_train):.2f}")
print(classification_report(y_train, y_pred_train))

# Test performance
y_pred_test = clf.predict(X_test_reduced)
print(f"\nTest Accuracy: {accuracy_score(y_test, y_pred_test):.2f}")
print(classification_report(y_test, y_pred_test))

#  PLOT NORMALIZED CONFUSION MATRIX 
def plot_cm(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, normalize='true')
    classes = np.unique(y_true)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt=".2f", cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

plot_cm(y_train, y_pred_train, "Training Confusion Matrix")
plot_cm(y_test,  y_pred_test,  "Test Set Confusion Matrix")
```

```python
import os
import random
import time

import numpy as np
import librosa
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#  REPRODUCIBILITY 
SEED = 42
np.random.seed(SEED)
random.seed(SEED)

#  PARAMETERS 
TESS_DIR    = "./TESSdataset"
RAVDESS_DIR = "./ravdess_by_emotion"
SR          = 16000
N_MFCC      = 13
MAX_LEN     = 85

emotion_map = {
    'angry': 0,
    'disgust': 1,
    'fear': 2,
    'happy': 3,
    'neutral': 4,
    'pleasant_surprised': 5,
    'sad': 6
}

#  FEATURE EXTRACTION 
def extract_mfcc(path):
    y, _ = librosa.load(path, sr=SR)
    hop = int(len(y) / (MAX_LEN - 1))
    mfcc = librosa.feature.mfcc(y=y, sr=SR, n_mfcc=N_MFCC, hop_length=hop)
    if mfcc.shape[1] < MAX_LEN:
        mfcc = np.pad(mfcc, ((0,0),(0,MAX_LEN-mfcc.shape[1])), mode='constant')
    else:
        mfcc = mfcc[:, :MAX_LEN]
    return mfcc

def load_raw_mfcc(base_dir):
    X, y = [], []
    for emo, label in emotion_map.items():
        emo_dir = os.path.join(base_dir, emo)
        if not os.path.isdir(emo_dir):
            continue
        for fname in os.listdir(emo_dir):
            if not fname.lower().endswith('.wav'):
                continue
            mf = extract_mfcc(os.path.join(emo_dir, fname))
            X.append(mf)
            y.append(label)
    return np.array(X), np.array(y)

#  LOAD & SPLIT 
X1, y1 = load_raw_mfcc(TESS_DIR)
X2, y2 = load_raw_mfcc(RAVDESS_DIR)
X_raw = np.vstack([X1, X2])
y_raw = np.hstack([y1, y2])

X_temp, X_test, y_temp, y_test = train_test_split(
    X_raw, y_raw, test_size=0.15, stratify=y_raw, random_state=SEED
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=SEED
)

#  FLATTEN FOR PCA/NMF 
Xtr = X_train.reshape(len(X_train), -1)
Xv  = X_val.reshape(len(X_val),   -1)
Xt  = X_test.reshape(len(X_test), -1)

#  TWO PCAs  8 DIM 
pca1 = PCA(n_components=4, random_state=SEED)
pca2 = PCA(n_components=4, random_state=0)

p1_tr = pca1.fit_transform(Xtr)
p1_v  = pca1.transform(Xv)
p1_t  = pca1.transform(Xt)

p2_tr = pca2.fit_transform(Xtr)
p2_v  = pca2.transform(Xv)
p2_t  = pca2.transform(Xt)

#  TWO NMFs  8 DIM 
nmf1 = NMF(n_components=4, init='random', random_state=SEED, max_iter=500)
nmf2 = NMF(n_components=4, init='nndsvda', random_state=0,     max_iter=500)

n1_tr = nmf1.fit_transform(np.abs(Xtr))
n1_v  = nmf1.transform(np.abs(Xv))
n1_t  = nmf1.transform(np.abs(Xt))

n2_tr = nmf2.fit_transform(np.abs(Xtr))
n2_v  = nmf2.transform(np.abs(Xv))
n2_t  = nmf2.transform(np.abs(Xt))

#  CONCAT  16D FEATURES 
feat_tr = np.hstack([p1_tr, p2_tr, n1_tr, n2_tr])
feat_v  = np.hstack([p1_v,  p2_v,  n1_v,  n2_v ])
feat_t  = np.hstack([p1_t,  p2_t,  n1_t,  n2_t ])

#  SCALE FEATURES 
scaler = StandardScaler()
feat_tr = scaler.fit_transform(feat_tr)
feat_v  = scaler.transform(feat_v)
feat_t  = scaler.transform(feat_t)

print("Feature shapes:", feat_tr.shape, feat_v.shape, feat_t.shape)
#  (n_samples, 16) each

#  TRAIN SVM 
svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=SEED)
svm.fit(feat_tr, y_train)

#  EVALUATE 
y_pred_val = svm.predict(feat_v)
y_pred_test = svm.predict(feat_t)

print(f"Validation Accuracy: {accuracy_score(y_val, y_pred_val):.4f}")
print(f"Test Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}\n")

print("Test Classification Report:")
print(classification_report(y_test, y_pred_test, target_names=list(emotion_map.keys())))

#  CONFUSION MATRIX 
cm = confusion_matrix(y_test, y_pred_test)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(8,6))
sns.heatmap(cm_norm, annot=True, fmt=".2f",
            xticklabels=emotion_map.keys(),
            yticklabels=emotion_map.keys(),
            cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Normalized Confusion Matrix (SVM)")
plt.show()
```
