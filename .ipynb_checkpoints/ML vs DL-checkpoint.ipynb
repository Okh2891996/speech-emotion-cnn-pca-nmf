```python
import os
import random
import numpy as np
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.svm          import SVC
from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics      import accuracy_score
from sklearn.model_selection import train_test_split

from tensorflow.keras import layers, regularizers

# ── REPRODUCIBILITY ──────────────────────────────────────────────────────────
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ── PATHS & PARAMETERS ───────────────────────────────────────────────────────
tess_dir    = "./TESSdataset"
ravdess_dir = "./ravdess_by_emotion"

# SVM MFCC settings
SAMPLE_RATE   = 8000
DURATION      = 0.5
TARGET_LENGTH = int(SAMPLE_RATE * DURATION)
N_MFCC        = 13  

# CNN MFCC‐map settings
sr         = 16000
n_mfcc     = 13
max_length = 85

# Training loop settings
TRAIN_FRACS = np.arange(0.10, 1.001, 0.05)  # 10%, 15%, …, 100%
EPOCHS      = 100
BATCH_SIZE  = 32

# Your 7‑emotion map for CNN labels
emotion_map = {
    'angry':               0,
    'disgust':             1,
    'fear':                2,
    'happy':               3,
    'neutral':             4,
    'pleasant_surprised':  5,
    'sad':                 6
}

# ── FEATURE EXTRACTION FUNCTIONS ─────────────────────────────────────────────
def extract_mfcc_svm(signal):
    mfccs = librosa.feature.mfcc(y=signal, sr=SAMPLE_RATE, n_mfcc=N_MFCC)
    feats = []
    for coeff in mfccs:
        feats += [np.mean(coeff), np.std(coeff)]
    return np.array(feats)

def process_directory_svm(directory):
    X, y = [], []
    for emo in sorted(os.listdir(directory)):
        emo_dir = os.path.join(directory, emo)
        if not os.path.isdir(emo_dir): continue
        for fname in sorted(os.listdir(emo_dir)):
            path = os.path.join(emo_dir, fname)
            try:
                sig, _ = librosa.load(path, sr=SAMPLE_RATE)
                sig = librosa.util.fix_length(sig, size=TARGET_LENGTH)
                sig /= (np.max(np.abs(sig))+1e-8)
                feats = extract_mfcc_svm(sig)
                X.append(feats)
                y.append(emo.lower())
            except:
                pass
    return np.vstack(X), np.array(y)

def extract_mfcc_cnn(file_path):
    y, _ = librosa.load(file_path, sr=sr)
    hop = int(len(y)/(max_length-1))
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, hop_length=hop)
    if mfcc.shape[1] < max_length:
        mfcc = np.pad(mfcc, ((0,0),(0,max_length-mfcc.shape[1])), mode='constant')
    else:
        mfcc = mfcc[:,:max_length]
    return np.expand_dims(mfcc, -1)

def load_all_data_cnn(base_dir):
    X, y = [], []
    for emo in sorted(os.listdir(base_dir)):
        p = os.path.join(base_dir, emo)
        if not os.path.isdir(p): continue
        lbl = emotion_map.get(emo.lower())
        if lbl is None: continue
        for f in sorted(os.listdir(p)):
            if not f.lower().endswith('.wav'): continue
            X.append(extract_mfcc_cnn(os.path.join(p,f)))
            y.append(lbl)
    return np.array(X), np.array(y)

# ── LOAD & PREPARE ────────────────────────────────────────────────────────────
# SVM data
X_svm_tess, y_svm_tess = process_directory_svm(tess_dir)
X_svm_rav,  y_svm_rav  = process_directory_svm(ravdess_dir)
X_svm = np.vstack((X_svm_tess, X_svm_rav))
y_svm = np.array([emotion_map[e] for e in np.concatenate((y_svm_tess, y_svm_rav))])

# CNN data
X_cnn_tess, y_cnn_tess = load_all_data_cnn(tess_dir)
X_cnn_rav,  y_cnn_rav  = load_all_data_cnn(ravdess_dir)
X_cnn = np.concatenate((X_cnn_tess, X_cnn_rav), axis=0)
y_cnn = np.concatenate((y_cnn_tess,  y_cnn_rav), axis=0)


# ── LOOP OVER TRAINING FRACTIONS ─────────────────────────────────────────────
# build fractions from 0.10 up to (but not including) 1.0, then append exactly 1.0
TRAIN_FRACS = list(np.round(np.arange(0.10, 1.0, 0.05), 2)) + [1.0]

svm_scores = []
cnn_scores = []

for frac in TRAIN_FRACS:
    print(f"→ Training on {int(frac*100)}% of data")

    # --- SVM subset & train ---
    if frac < 1.0:
        X_sub_svm, _, y_sub_svm, _ = train_test_split(
            X_svm_train, y_svm_train,
            train_size=frac,
            stratify=y_svm_train,
            random_state=SEED
        )
    else:
        # use all training data
        X_sub_svm, y_sub_svm = X_svm_train, y_svm_train

    # scale & fit
    scaler = StandardScaler()
    X_sub_svm_scaled = scaler.fit_transform(X_sub_svm)
    X_svm_test_scaled = scaler.transform(X_svm_test)

    svm = SVC(kernel='rbf', C=10, gamma='scale', random_state=SEED)
    svm.fit(X_sub_svm_scaled, y_sub_svm)
    y_pred_svm = svm.predict(X_svm_test_scaled)
    svm_scores.append(accuracy_score(y_svm_test, y_pred_svm))

    # --- CNN subset & train ---
    if frac < 1.0:
        X_sub_cnn, _, y_sub_cnn, _ = train_test_split(
            X_cnn_train, y_cnn_train,
            train_size=frac,
            stratify=y_cnn_train,
            random_state=SEED
        )
    else:
        X_sub_cnn, y_sub_cnn = X_cnn_train, y_cnn_train

    # build & compile fresh model
    cnn_model = EmotionClassifier(len(emotion_map))
    cnn_model.build(input_shape=(None, n_mfcc, max_length, 1))
    cnn_model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    cnn_model.fit(
        X_sub_cnn, y_sub_cnn,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=0
    )
    _, cnn_acc = cnn_model.evaluate(X_cnn_test, y_cnn_test, verbose=0)
    cnn_scores.append(cnn_acc)

# ── PLOT RESULTS ─────────────────────────────────────────────────────────────
plt.figure(figsize=(8,5))
plt.plot(np.array(TRAIN_FRACS)*100, svm_scores, '-o', label='SVM (ML)')
plt.plot(np.array(TRAIN_FRACS)*100, cnn_scores, '-o', label='CNN (DL)')
plt.xlabel('Percentage of Training Data (%)')
plt.ylabel('Test Accuracy')
plt.title('ML vs DL Performance as Training Data Grows')
plt.legend()
plt.grid(True)
plt.show()

```
