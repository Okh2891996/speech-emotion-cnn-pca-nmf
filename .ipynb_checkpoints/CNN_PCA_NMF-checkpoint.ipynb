# NMF = 2, PCA = 2 

```python
import os
import random
import time

import numpy as np
import librosa
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA, NMF
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.callbacks import CSVLogger

import tensorflow as tf
from tensorflow.keras import layers, Model, regularizers

# ── REPRODUCIBILITY ─────────────────────────────────────────────────────────
SEED = 42
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ── PARAMETERS ──────────────────────────────────────────────────────────────
TESS_DIR    = "./TESSdataset"
RAVDESS_DIR = "./ravdess_by_emotion"
SR          = 16000
N_MFCC      = 13
MAX_LEN     = 85
NUM_CLASSES = 7

emotion_map = {
    'angry': 0, 'disgust': 1, 'fear': 2, 'happy': 3,
    'neutral': 4, 'pleasant_surprised': 5, 'sad': 6
}

# ── FEATURE EXTRACTION ───────────────────────────────────────────────────────
def extract_mfcc(path):
    y, _ = librosa.load(path, sr=SR)
    hop = int(len(y) / (MAX_LEN - 1))
    mfcc = librosa.feature.mfcc(
        y=y, sr=SR, n_mfcc=N_MFCC, hop_length=hop
    )
    if mfcc.shape[1] < MAX_LEN:
        mfcc = np.pad(mfcc, ((0,0),(0, MAX_LEN-mfcc.shape[1])), mode='constant')
    else:
        mfcc = mfcc[:, :MAX_LEN]
    return mfcc

def load_dataset(base_dir):
    X, y = [], []
    for emo, lbl in emotion_map.items():
        emo_dir = os.path.join(base_dir, emo)
        if not os.path.isdir(emo_dir):
            continue
        for fname in os.listdir(emo_dir):
            if not fname.lower().endswith('.wav'):
                continue
            X.append(extract_mfcc(os.path.join(emo_dir, fname)))
            y.append(lbl)
    return np.array(X), np.array(y)

# ── LOAD & SPLIT ─────────────────────────────────────────────────────────────
X1, y1 = load_dataset(TESS_DIR)
X2, y2 = load_dataset(RAVDESS_DIR)
X_raw = np.vstack([X1, X2])
y_raw = np.hstack([y1, y2])

X_temp, X_test, y_temp, y_test = train_test_split(
    X_raw, y_raw, test_size=0.15, stratify=y_raw, random_state=SEED
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=SEED
)

# ── FLATTEN FOR PCA/NMF ───────────────────────────────────────────────────────
Xtr = X_train.reshape(len(X_train), -1)
Xv  = X_val.reshape(len(X_val),   -1)
Xt  = X_test.reshape(len(X_test), -1)

# ── TWO PCA’s → 2+2 dims ─────────────────────────────────────────────────────
pca1 = PCA(n_components=2, random_state=SEED)
pca2 = PCA(n_components=2, random_state=0)

p1_tr, p1_v,  p1_t  = pca1.fit_transform(Xtr), pca1.transform(Xv),  pca1.transform(Xt)
p2_tr, p2_v,  p2_t  = pca2.fit_transform(Xtr), pca2.transform(Xv),  pca2.transform(Xt)

# ── TWO NMF’s → 2+2 dims ─────────────────────────────────────────────────────
nmf1 = NMF(n_components=2, init='random',    random_state=SEED, max_iter=500)
nmf2 = NMF(n_components=2, init='nndsvda',   random_state=0,   max_iter=500)

n1_tr = nmf1.fit_transform(np.abs(Xtr)); n1_v = nmf1.transform(np.abs(Xv)); n1_t = nmf1.transform(np.abs(Xt))
n2_tr = nmf2.fit_transform(np.abs(Xtr)); n2_v = nmf2.transform(np.abs(Xv)); n2_t = nmf2.transform(np.abs(Xt))

# ── CONCAT & SCALE → 8‑D FEATURES ────────────────────────────────────────────
feat_tr = np.hstack([p1_tr, p2_tr, n1_tr, n2_tr])
feat_v  = np.hstack([p1_v,  p2_v,  n1_v,  n2_v ])
feat_t  = np.hstack([p1_t,  p2_t,  n1_t,  n2_t ])

scaler = StandardScaler()
feat_tr = scaler.fit_transform(feat_tr)
feat_v  = scaler.transform(feat_v)
feat_t  = scaler.transform(feat_t)

print("Feature shapes:", feat_tr.shape, feat_v.shape, feat_t.shape)
# → (n_samples, 8) each

# ── PREPARE CNN INPUT + AUGMENT ─────────────────────────────────────────────
X_train = X_train[..., np.newaxis]
X_val   = X_val[...,   np.newaxis]
X_test  = X_test[...,  np.newaxis]

noise_std = 0.01
X_noisy   = X_train + noise_std * np.random.randn(*X_train.shape)
X_aug     = np.vstack([X_train, X_noisy])
F_aug     = np.vstack([feat_tr, feat_tr])
y_aug     = np.hstack([y_train, y_train])

perm = np.random.permutation(len(X_aug))
X_aug, F_aug, y_aug = X_aug[perm], F_aug[perm], y_aug[perm]

# ── BUILD MODEL ──────────────────────────────────────────────────────────────
mfcc_in = tf.keras.Input(shape=(N_MFCC, MAX_LEN, 1), name='mfcc_input')
feat_in = tf.keras.Input(shape=(8,), name='feat_input')

x = mfcc_in
for filters in [36, 72, 144]:
    x = layers.Conv2D(filters, 3, padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(2)(x)
    x = layers.Dropout(0.3)(x)
x = layers.Flatten()(x)

merged = layers.Concatenate()([x, feat_in])
h = layers.Dense(256, activation='relu',
                 kernel_regularizer=regularizers.l2(1e-4))(merged)
h = layers.Dropout(0.4)(h)
out = layers.Dense(NUM_CLASSES, activation='softmax')(h)

model = Model([mfcc_in, feat_in], out)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# ── TRAIN ────────────────────────────────────────────────────────────────────
csv_logger = CSVLogger('training.log', append=False)
start = time.time()
history = model.fit(
    [X_aug, F_aug], y_aug,
    validation_data=([X_val, feat_v], y_val),
    epochs=50,
    batch_size=32,
    shuffle=False,       # already shuffled above
    callbacks=[csv_logger]
)
print(f"Training time: {time.time()-start:.1f}s")

# ── EVALUATE ─────────────────────────────────────────────────────────────────
loss, acc = model.evaluate([X_test, feat_t], y_test, verbose=0)
print(f"\nTest loss: {loss:.4f},  Test accuracy: {acc:.4f}")

y_pred = np.argmax(model.predict([X_test, feat_t]), axis=1)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=list(emotion_map.keys())))

cm = confusion_matrix(y_test, y_pred)
cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)
plt.figure(figsize=(8,6))
sns.heatmap(cm_norm, annot=True, fmt=".2f",
            xticklabels=emotion_map.keys(),
            yticklabels=emotion_map.keys(),
            cmap="Blues")
plt.xlabel("Predicted"); plt.ylabel("True")
plt.title("Normalized Confusion Matrix")
plt.show()

# ── PLOT TRAINING HISTORY ─────────────────────────────────────────────────────
plt.figure(figsize=(8,4))
plt.plot(history.history['accuracy'],  label='Train Acc')
plt.plot(history.history['val_accuracy'],label='Val Acc')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Accuracy')
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(history.history['loss'],      label='Train Loss')
plt.plot(history.history['val_loss'],  label='Val Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend(); plt.grid(True)
plt.show()
```
